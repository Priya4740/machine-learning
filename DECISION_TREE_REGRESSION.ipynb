```python
x=pd.DataFrame(df['rm'])
y=pd.DataFrame(df['medv'])
```

```python
x=pd.DataFrame(df,columns=['crim','zn','indus','chas','nox','rm','age','dis','rad','tax','ptratio','black','lstat'])
y=pd.DataFrame(df,columns=['medv'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
y=df["medv"]
```

```python
df.shape
```

```python
from sklearn.tree import DecisionTreeRegressor
Regressor=DecisionTreeRegressor(max_depth=5,random_state=42)
Regressor.fit(x_train,y_train)
print("Model Trained Successfully")
```

```python
y_pred=Regressor.predict(x_test)
comparsion=pd.DataFrame({"Actual":y_test.values.ravel(),"Predicted":y_pred})
comparsion.head(10)
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_pred)
rmse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
rmse
```

```python
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(y_test,y_pred)
print("MAE:",mae)
```

```python
from sklearn import metrics
print("R^2 Score:",metrics.r2_score(y_test,y_pred))
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
plot_tree(Regressor,feature_names=x.columns,class_names=["lstat","medv"],filled=True,rounded=True,fontsize=10)
plt.show()
```
