```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Creating sample Dataset


```python
data={
    "Area":[1000,1200,1500,1800,2000,2500],
    "Price":[10,12,15,18,20,25]
}
df=pd.DataFrame(data)
df
```

Visualizing the dataset

```python
plt.scatter(df["Area"],df["Price"])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("House Area vs Price")
plt.show()
```

Building linear regression model

```python
from sklearn.linear_model import LinearRegression
```

```python
x=df[["Area"]]
y=df["Price"]
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

Predicting HousePrice

```python
prediction=model.predict([[2200]])
print("Predicted House Price:",prediction[0],"Lakhs")
```

Visualizing Regression Line

```python
plt.scatter(df["Area"],df["Price"],color="blue")
plt.plot(df["Area"],model.predict(x),color="red")
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Linear Regression Line")
plt.show()
```

Model Coefficient & Intercept

```python
print("slope(m):",model.coef_[0])
print("Intercept(c):",model.intercept_)
```
